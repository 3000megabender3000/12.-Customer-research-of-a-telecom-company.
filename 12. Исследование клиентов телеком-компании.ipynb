## Описание проекта:
<a id='01'></a>

Проект представляет собой исследование клиентов телеком-компании «Нупозвони», которая является провайдером виртуальной телефонии. Клиенты компании - это колл-центры, которые распределяют входящие вызовы по своим операторам и обеспечивают исходящие вызовы от операторов. Предполагается, что клиенты с неоптимальными расходами могут уйти к конкурентам. В проекте проводится мониторинг активности пользователей, их расходы с целью выявления неоптимальных затрат. В дальнейшем, для таких клиентов будут разработаны предложения по снижению затрат путем перехода на более выгодный тариф. Паралллельно будет оцениваться возможное снижение выручки при полном или частичном переводе клиентов с неоптимальными тарифами на более выгодные тарифы. Заказчиком проекта является коммерческий директор. Итоги проекта предназначены для коммерческого директора и коммерческого отдела, консультанты которого будут обзванивать клиентов с предложением о переходе на более выгодный тариф.
Разработанные в проекте метрики оценки клиентской активности и расходов будут использоваться на постоянной основе.

Создам функцию для подсчета пропущенных значений и дубликатов.

```python
def missing_duplicated_values(data):
        '''Функция получает на вход исходный датасет и подсчитывает и выводит на экран таблицу 
        количества пропущенных сообщений missing_values и долю таких сообщений % of total в общем числе, 
        а также количество полных дубликатов. 
        Используются методы библиотеки pandas - isna и duplicated. 
        '''
        missing = data.isna().sum().to_frame()
        missing = missing.rename(columns = {0:'missing_values'})
        missing['% of total'] = ((missing['missing_values'] / data.shape[0]) * 100).round(2)
        missing_sum = missing['missing_values'].sum()
        duplicated_sum = data.duplicated().sum()
        display (missing)
        print('Найдено пропущенных значений: {}, что составляет {}% от общего числа'
              .format(missing_sum, ((missing_sum/data.shape[0])*100).round(2)))
        print('Найдено дубликатов: {}, что составляет {}% от общего числа'
              .format(duplicated_sum, ((duplicated_sum/data.shape[0])*100).round(2)))
```

Создам функцию costs, которая рассчитывает расход клиента в зависимости от тарифа. Перед этим добавлю тарифные ставки в таблицу. Создам столбец fee с абонплатой, столбец over_in со ставкой по перерасходу минут на исходящие внутри сети и столбец over_out со ставкой по перерасходу минут на исходящие вне сети.

```python
df['fee'] = df['tariff_plan'].replace({'A':5000, 'B':2000, 'C':1000})
df['over_in'] = df['tariff_plan'].replace({'A':0.1, 'B':0.15, 'C':0.3})
df['over_out'] = df['tariff_plan'].replace({'A':0.4, 'B':0.5, 'C':0.7})
df['date'] = pd.to_datetime(df['date']).dt.strftime("%Y-%m")
df.head()
```

```python
def costs(user_id):
    '''Функция получает на вход строку с user_id. Далее делается срез данных по исходящим звонкам, и
    затем, перебирая в цикле месяцы расчетного периода, рассчитывается суммарный по всем месяцам 
    расход клиента и выводится в виде числа.    
    '''
    #data = df.copy(deep = True)
    data = data1.copy(deep = True)
    months = data['date'].unique().tolist() # Создаю список расчетных месяцев
    data = data.query('user_id == @user_id') # Ограничиваю датасет одним пользователем
    dur_in = data.iloc[0,12] #Стоимость минуты перерасхода исходящего внутри сети
    dur_out = data.iloc[0,13] #Стоимость минуты перерасхода исходящего вне сети
    cost = 0
    # Расчет количества операторов и платы за них
    operators_cost = data.query('operator_id != "not defined"').groupby('date')['operator_id'].nunique().sum() * 100
    fee_all = data.iloc[0,11] * len(months) # Расчет тарифной платы за весь период
    for month in months:
        dfq = data.query('date == @month')
        if dfq.shape[0] == 0:
            duration_in = 0
            duration_out = 0
        else:
            dfq = dfq.query('direction == "out"')
            duration_in  = np.ceil((dfq.query('internal == True')['total_call_duration'].sum())/60)
            if (duration_in / 60) > 2000: duration_in = (duration_in - 2000) * dur_in
            else: duration_in = 0
            duration_out = np.ceil(dfq.query('internal == False')['total_call_duration'].sum()/60) * dur_out
        cost_month = duration_in + duration_out
        cost += cost_month 
    cost = cost + fee_all + operators_cost
    return cost
```

Создам функцию company_income, которая считает доходы компании  по тарифам.

```python
def company_income(data0):
    '''Функция получает на вход таблицу с исходными данными об активности. 
    Получающийся результат расчета прибыли компании объявляется глобальной переменной (доступной вне функции). Сначала 
    функция группирует данные по user_id и считает сумму total_call_duration. Далее построчно применяется функция costs
    для расчета прибыли компании по каждому тарифу. Расчеты сохраняются в таблицу df_costs.
    '''
    global data1
    data1 = data0.copy(deep = True)
    df_costs = data0.copy(deep = True)
    df_costs = data0.groupby(['user_id', 'tariff_plan'], as_index = False).agg({'total_call_duration' : 'sum'})
    df_costs['costs'] = df_costs['user_id'].apply(costs)
    revenue = df_costs['costs'].sum().round(2)
    print('Общая выручка компании за весь период составляет: {} рублей'.format(revenue))
    return revenue, df_costs
```

Создам функцию company_income_visual для визуализации распределения доходов компании.

```python
def company_income_visual(data_costs):
    '''На экран выводится piechart распределения доходов компании по тарифам. 
    Для вывода используется библиотека matplotlib и модуль pyplot.'''
    costs_by_tariff = data_costs.groupby('tariff_plan', as_index = False).agg({'costs':'sum'})
    plt.style.use('seaborn-pastel')
    textprops = {"fontsize":14}
    plt.figure(figsize=(16,9))
    plt.pie(costs_by_tariff.costs, labels = costs_by_tariff['tariff_plan'], autopct = "%0.2f%%", startangle = 90, textprops = textprops, wedgeprops = {"linewidth": 1, 'width':0.7, "edgecolor":"k"})
    plt.title('Распределение доходов компании по тарифам', fontdict = {'fontsize' : 20})
    plt.legend()
    plt.show()
```

## Статистический анализ данных: 
<a id='05'></a>   
<li>проверить гипотезу о равенстве средних для зависимых (парных) выборок - взять выборки выручек с клиентов на имеющихся тарифах и выручки после перевода клиентов на более выгодные тарифы 
<li>гипотеза - изменится ли выручка компании после перехода клиентов на выгодные тарифы
</ul>

Чтобы проверить гипотезу о равенстве средних двух генеральных совокупностей для зависимых (парных) выборок в Python, применим метод scipy.stats.ttest_rel(). Достаточно передать всего два параметра: массивы с данными до и после. Размеры массивов должны быть одинаковыми.

Парная выборка означает, что мы измеряем некоторую переменную для одних и тех же единиц. В нашем распоряжении расходы для каждого пользователя — до и после изменения.

Сформулирую нулевую и альтернативную гипотезу.
Нулевая гипотеза: выручка компании после перевода клиентов на выгодные тарифы не изменится.
Альтернативная гипотеза: выручка изменится.
Критический уровень статистической значимости alpha возьму равный 0.05.
Создам функцию income_change для оценки изменения выручки.

```python
def income_change(data_with_costs):
    '''Это двусторонний тест на нулевую гипотезу о том, 
    что 2 независимые выборки имеют одинаковые средние (ожидаемые) значения. 
    '''
    alpha = 0.05
    before = data_with_costs['costs_own']
    after = data_with_costs['min_costs']
    
    #results = st.ttest_rel(before, after)
    results = st.ttest_ind(before, after)
    
    print('p-значение: ', results.pvalue)

    if results.pvalue < alpha:
        print("Отвергаем нулевую гипотезу, выручка компании изменилась")
    else:
        print("Не получилось отвергнуть нулевую гипотезу, выручка компании не изменилась") 
```

```python
income_change(df_costs_pivot)
```

<b>При переводе клиентов на более выгодные для них тарифы, изменение доходов компании будет превышать уровень статистической значимости. То есть изменение выручки будет существенным.</b>
